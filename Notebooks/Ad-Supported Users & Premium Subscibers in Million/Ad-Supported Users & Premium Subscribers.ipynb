Ratio of Ad-Supported Users vs. Premium Subscibers from 2019 - 2024 in Million:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read CSV, skipping the first row since it's just a title
df = pd.read_csv("Ad-Supported Users & Premium Subscibers in Million.csv", skiprows=1)

# Clean column names (remove spaces)
df.columns = [col.strip() for col in df.columns]

# Remove "M" and convert values to integers
df["Ad-Supported Users (Million)"] = df["Ad-Supported Users (Million)"].str.replace("M", "").astype(int)
df["Premium Subscribers (Million)"] = df["Premium Subscribers (Million)"].str.replace("M", "").astype(int)

# Set figure size
plt.figure(figsize=(12, 6))

# Plot the lines with custom colors
sns.lineplot(x=df["Year"], y=df["Ad-Supported Users (Million)"], label="Ad-Supported Users", marker="o", color="DarkTurquoise")
sns.lineplot(x=df["Year"], y=df["Premium Subscribers (Million)"], label="Premium Subscribers", marker="s", color="purple")

# Add values on top of each point
for i in range(len(df)):
    plt.text(df["Year"][i], df["Ad-Supported Users (Million)"][i] + 20, str(df["Ad-Supported Users (Million)"][i]), 
             color="black", ha="center", fontsize=10, fontweight="bold")
    plt.text(df["Year"][i], df["Premium Subscribers (Million)"][i] + 20, str(df["Premium Subscribers (Million)"][i]), 
             color="black", ha="center", fontsize=10, fontweight="bold")

# Labels and title
plt.xlabel("Year")
plt.ylabel("Number of Users (Million)")
plt.title("Ad-Supported Users vs. Premium Subscribers (2019-2024) in Million")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
