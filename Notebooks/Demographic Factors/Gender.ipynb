Spotify's Gender Demographics

Sources are from a thirdparty Website (January 9th 2025)!

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file, skipping the first row to remove headers inside the data
df = pd.read_csv("Gender.csv", skiprows=1, names=["Gender", "Percentage"], dtype=str)

# Remove leading and trailing spaces
df = df.apply(lambda x: x.str.strip())

# Drop non-numeric rows and convert 'Percentage' to float
df = df[pd.to_numeric(df["Percentage"], errors="coerce").notna()]
df["Percentage"] = df["Percentage"].astype(float)

# Values for the chart
genders = df["Gender"].tolist()
percentages = df["Percentage"].tolist()

# Colors for the chart
colors = ["#FF69B4", "#00b2ee"]

# Create donut chart
plt.figure(figsize=(8, 8))
wedges, _, autotexts = plt.pie(
    percentages, labels=[None] * len(genders), autopct="%1.1f%%", colors=colors, startangle=140, wedgeprops={"edgecolor": "black"}
)

# Add a white circle in the center to create a donut effect
center_circle = plt.Circle((0, 0), 0.70, fc="white")
plt.gca().add_artist(center_circle)

# Adjust percentage text size and color
for autotext in autotexts:
    autotext.set_fontsize(15)
    autotext.set_color("black")

# Title and legend
plt.title("Spotify's Gender Demographics in (%)")
plt.legend(wedges, genders, title="Gender", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Show the chart
plt.show()
```
