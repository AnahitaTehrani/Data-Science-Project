```python
import pandas as pd
import matplotlib.pyplot as plt

# Ad-Supported Revenue Pie Charts
def plot_ad_supported_revenue():
    df = pd.read_csv("Ad-Supported Revenue.csv", index_col=0)
    colors = {"Europe": "#1f77b4", "North America": "#ff7f0e", "Latin America": "#2ca02c", "Rest of World": "#d62728"}
    for year in df.index:
        plt.figure(figsize=(6, 6))
        plt.pie(df.loc[year], labels=df.columns, autopct='%1.1f%%', colors=[colors[region] for region in df.columns])
        plt.title(f"Ad-Supported Revenue - {year}")
        plt.legend(df.columns, loc="best")
        plt.show()

plot_ad_supported_revenue()

# Premium Revenue Pie Charts
def plot_premium_revenue():
    df = pd.read_csv("Premium Revenue.csv", index_col=0)
    colors = {"Europe": "#1f77b4", "North America": "#ff7f0e", "Latin America": "#2ca02c", "Rest of World": "#d62728"}
    for year in df.index:
        plt.figure(figsize=(6, 6))
        plt.pie(df.loc[year], labels=df.columns, autopct='%1.1f%%', colors=[colors[region] for region in df.columns])
        plt.title(f"Premium Revenue - {year}")
        plt.legend(df.columns, loc="best")
        plt.show()

plot_premium_revenue()

# Spotify Revenue Pie Charts
def plot_spotify_revenue():
    df = pd.read_csv("Spotify Revenue.csv", index_col=0)
    colors = {"Europe": "#1f77b4", "North America": "#ff7f0e", "Latin America": "#2ca02c", "Rest of World": "#d62728"}
    for year in df.index:
        plt.figure(figsize=(6, 6))
        plt.pie(df.loc[year], labels=df.columns, autopct='%1.1f%%', colors=[colors[region] for region in df.columns])
        plt.title(f"Spotify Revenue - {year}")
        plt.legend(df.columns, loc="best")
        plt.show()

plot_spotify_revenue()
```
